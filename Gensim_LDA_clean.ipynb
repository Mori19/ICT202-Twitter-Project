```python
import pandas as pd
```

```python
df = pd.read_json('post_process_data.json')
docs = df['content']
docs = [doc.split() for doc in docs.to_list()]
```

```python
import gensim 
import gensim.corpora as corpora
from gensim.models import CoherenceModel
from gensim.models.ldamodel import LdaModel
from gensim.models.ldamulticore import LdaMulticore
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
id2word = corpora.Dictionary(docs)
corpus = [id2word.doc2bow(text) for text in docs]
```

```python
coherence = []

for k in range(5,50,10):
    lda_model = LdaMulticore(corpus=corpus,
                        id2word=id2word,
                        num_topics=k,
                        random_state=1,
                        passes=25,
                        workers=10)
    coherence.append(CoherenceModel(model=lda_model, texts=docs, dictionary=id2word, coherence='c_v').get_coherence())
```

```python
plt.plot(range(5,50,10),coherence)
plt.title('Coherence Scores')
plt.xlabel('Topics')
plt.ylabel('Coherence')
plt.show()
```

```python
#need to tape my doc list to a its topic
topic_to_doc = {}
for i,doc in enumerate(docs):
    doc_topic = lda_model.get_document_topics(id2word.doc2bow(doc))
    doc_topic.sort(key=lambda y: y[1],reverse=True)
    topic_to_doc[i] = {'doc':doc,'topic':doc_topic[0][0],'likeliness':doc_topic[0][1]}
docs_df = pd.DataFrame(topic_to_doc).T
```

```python
docs_df.head()
```

```python
from gensim.models.doc2vec import TaggedDocument
from gensim.models.doc2vec import Doc2Vec
```

```python
def tag(doc_list):
    for i, doc in enumerate(doc_list):
        yield TaggedDocument(doc, [i])
```

```python
tagged_docs = list(tag(docs_df['doc']))
```

```python
tagged_docs[:5]
```

```python
model = Doc2Vec(vector_size=50, min_count=5, epochs=40)
model.build_vocab(tagged_docs)
```

```python
model.train(tagged_docs, total_examples=model.corpus_count, epochs=model.epochs)
```

```python
X = model.dv[range(len(model.dv))]
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
pca = PCA(n_components=2)

result = pca.fit_transform(X)
pca_df = pd.DataFrame(result, columns = ['x','y'])
plt.scatter(pca_df['x'], pca_df['y'])
plt.show()
```

```python
joint_df = docs_df.join(pca_df)
```
